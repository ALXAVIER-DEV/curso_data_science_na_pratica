# Gráfico de Dispersão (*Scatter Plot*)

O gráfico de dispersão (*scatter plot*) é um gráfico bi-dimensional para visualização de dados que usa pontos para representar o valor entre duas variáveis diferentes.

Uma das variáveis corresponde ao valor $x$ e a outra variável corresponde ao valor $y$.

Esse tipo de gráfico é ideal quando você quer mostrar a relação entre duas variáveis, e dá ótimos indícios sobre a correlação existente entre elas.


<center><img src="https://www.data-to-viz.com/graph/scatter_files/figure-html/unnamed-chunk-1-1.png" height="300px"></center>

Vale ressaltar que nem todas as relações são lineares. Muitas delas seguem outros tipos de padrões não-lineares ou não seguem nenhum  tipo de padrão aparente.

## Plotar o *Scatter Plot*


```
# importar os pacotes necessários
import pandas as pd
import matplotlib.pyplot as plt

# importar o dataset
df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/data_science/master/datasets/housing.csv")

# ver as primeiras entradas
df.head()
```

```
# plotar o scatterplot
fig, ax = plt.subplots(figsize=(10,6))

df.plot.scatter(x='median_income', y='median_house_value', ax=ax, s=1)

plt.tight_layout()
```

## Regressão Linear em um *Scatter Plot*

Só para efeito de curiosidade e conhecimento, em um gráfico de dispersão é possível achar uma reta que melhor represente todos aqueles pontos plotados. Veja como plotar uma reta por meio da Regressão Linear usando a biblioteca `seaborn` e o método `sns.regplot()`.

```
# importar o seaborn
import seaborn as sns

# plotar scatter plot e regressão linear
fig, ax = plt.subplots()

ax.set_xlim(right=8)
sns.regplot('median_income', 'median_house_value', scatter_kws={"s": 0.2},
            line_kws={"color": "red"}, data=df, ax=ax)

plt.tight_layout()
```
